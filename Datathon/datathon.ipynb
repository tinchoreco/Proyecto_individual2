```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
pd.options.display.float_format = '{:,.2f}'.format
```

```python
data_train = pd.read_csv('properties_colombia_train.csv')
data_test = pd.read_csv('properties_colombia_test.csv')
```

```python
promedio = data_train.price.mean()

data_train['Target'] = data_train['price'].apply(lambda x:1 if x>promedio else 0)
data_train.head()
```

Completamos los valores faltantes con el valor promedio en Latitud, Longitud y Numero de Baños

```python
# Con el Data Test
promedio_lat = data_train.lat.mean()
promedio_lon = data_train.lon.mean()
promedio_bathroom = data_train.bathrooms.mean()

data_train['lat'] = data_train['lat'].fillna(value=promedio_lat)
data_train['lon'] = data_train['lon'].fillna(value=promedio_lon)
data_train['bathrooms'] = data_train['bathrooms'].fillna(value=promedio_bathroom)

# Con el Data Train
promedio_lat = data_test.lat.mean()
promedio_lon = data_test.lon.mean()
promedio_bathroom = data_test.bathrooms.mean()

data_test['lat'] = data_test['lat'].fillna(value=promedio_lat)
data_test['lon'] = data_test['lon'].fillna(value=promedio_lon)
data_test['bathrooms'] = data_test['bathrooms'].fillna(value=promedio_bathroom)

data_test.head()
```

Quitamos las columnas que no utilizamos

```python
# Data Train
data_train.drop('id', axis= 1, inplace= True)
data_train.drop('ad_type', axis= 1, inplace= True)
data_train.drop('start_date', axis= 1, inplace= True)
data_train.drop('end_date', axis= 1, inplace= True)
data_train.drop('created_on', axis= 1, inplace= True)
data_train.drop('surface_total', axis= 1, inplace= True)
data_train.drop('surface_covered', axis= 1, inplace= True)
data_train.drop('rooms', axis= 1, inplace= True)
data_train.drop('bedrooms', axis= 1, inplace= True)
data_train.drop('price_period', axis= 1, inplace= True)
data_train.drop('title', axis= 1, inplace= True)
data_train.drop('description', axis= 1, inplace= True)
data_train.drop('operation_type', axis= 1, inplace= True)
data_train.drop('l1', axis= 1, inplace= True)
data_train.drop('l4', axis= 1, inplace= True)
data_train.drop('l5', axis= 1, inplace= True)
data_train.drop('l6', axis= 1, inplace= True)
data_train.drop('geometry', axis=1, inplace= True)

# Data Train
data_test.drop('id', axis= 1, inplace= True)
data_test.drop('ad_type', axis= 1, inplace= True)
data_test.drop('start_date', axis= 1, inplace= True)
data_test.drop('end_date', axis= 1, inplace= True)
data_test.drop('created_on', axis= 1, inplace= True)
data_test.drop('surface_total', axis= 1, inplace= True)
data_test.drop('surface_covered', axis= 1, inplace= True)
data_test.drop('rooms', axis= 1, inplace= True)
data_test.drop('bedrooms', axis= 1, inplace= True)
data_test.drop('price_period', axis= 1, inplace= True)
data_test.drop('title', axis= 1, inplace= True)
data_test.drop('description', axis= 1, inplace= True)
data_test.drop('operation_type', axis= 1, inplace= True)
data_test.drop('l1', axis= 1, inplace= True)
data_test.drop('l4', axis= 1, inplace= True)
data_test.drop('l5', axis= 1, inplace= True)
data_test.drop('l6', axis= 1, inplace= True)
data_test.drop('geometry', axis=1, inplace= True)

```

Cambiamos las columnas de tipo object a Category 

```python
from sklearn.preprocessing import OneHotEncoder
# Data Train
data_train['l2'] = data_train['l2'].astype("category")
data_train['l3'] = data_train['l3'].astype("category")
data_train['property_type'] = data_train['property_type'].astype("category")

# codificamos la columnas l2 de datos categoricos con OneHot
codificador = OneHotEncoder()
codificacion = codificador.fit_transform(data_train[['l2']])
nuevas_cols = pd.DataFrame(codificacion.toarray(),columns= codificador.categories_)

data_train = pd.concat([data_train, nuevas_cols], axis=1)

# codificamos la columnas l3 de datos categoricos con OneHot
codificador = OneHotEncoder()
codificacion = codificador.fit_transform(data_train[['l3']])
nuevas_cols = pd.DataFrame(codificacion.toarray(),columns= codificador.categories_)

data_train = pd.concat([data_train, nuevas_cols], axis=1)

# codificamos la columnas property_type de datos categoricos con OneHot
codificador = OneHotEncoder()
codificacion = codificador.fit_transform(data_train[['property_type']])
nuevas_cols = pd.DataFrame(codificacion.toarray(),columns= codificador.categories_)

data_train = pd.concat([data_train, nuevas_cols], axis=1)

# finalmente eliminamos las columnas l2, l3 y property_type
data_train.drop('l2', axis= 1, inplace= True)
data_train.drop('l3', axis= 1, inplace= True)
data_train.drop('property_type', axis= 1, inplace= True)

# Data test
data_test['l2'] = data_test['l2'].astype("category")
data_test['l3'] = data_test['l3'].astype("category")
data_test['property_type'] = data_test['property_type'].astype("category")

# codificamos la columnas l2 de datos categoricos con OneHot
codificador = OneHotEncoder()
codificacion = codificador.fit_transform(data_test[['l2']])
nuevas_cols = pd.DataFrame(codificacion.toarray(),columns= codificador.categories_)

data_test = pd.concat([data_test, nuevas_cols], axis=1)

# codificamos la columnas l3 de datos categoricos con OneHot
codificador = OneHotEncoder()
codificacion = codificador.fit_transform(data_test[['l3']])
nuevas_cols = pd.DataFrame(codificacion.toarray(),columns= codificador.categories_)

data_test = pd.concat([data_test, nuevas_cols], axis=1)

# codificamos la columnas property_type de datos categoricos con OneHot
codificador = OneHotEncoder()
codificacion = codificador.fit_transform(data_test[['property_type']])
nuevas_cols = pd.DataFrame(codificacion.toarray(),columns= codificador.categories_)

data_test = pd.concat([data_test, nuevas_cols], axis=1)

# finalmente eliminamos las columnas l2, l3 y property_type
data_test.drop('l2', axis= 1, inplace= True)
data_test.drop('l3', axis= 1, inplace= True)
data_test.drop('property_type', axis= 1, inplace= True)


data_train.head()
```

```python
# Data Train
data_train.drop('currency', axis= 1, inplace= True)
data_train.drop('price', axis=1, inplace= True)

# Data Test
data_test.drop('currency', axis= 1, inplace= True)
data_train.head()
```

```python
X = data_train.copy()
X.drop(['Target', ('Albania',), ('Arcabuco',), ('Balboa',), ('Buenaventura',), ('Buesaco',), ('Carmen de Carupa',), ('Casabianca',), ('Choachí',), ('Circasia',), ('Coello',), ('Cunday',), ('Dos Quebradas',), ('Guateque',), ('Gómez Plata',), ('Lérida',), ('Nocaima',), ('Pacho',), ('Pasto',), ('Pinchote',), ('Pore',), ('Providencia',), ('Puerto Carreño',), ('Quebradanegra',), ('Riohacha',), ('San Alberto',), ('San Pedro de Uraba',), ('Santa Rita',), ('Santa Rosa',), ('Santa Sofía',), ('Santagueda',), ('Sutamarchán',), ('Sutatausa',), ('Suárez',), ('Tausa',), ('Vergara',), ('Vianí',)], axis=1, inplace=True)
y = data_train.Target

X_test = data_test
```

## **1. Árbol de decisión**

Ahora que tenemos nuestras variables predictoras `X`, por un lado, y nuestra varible a predecir `y`, por el otro, vamos a crear un modelo de árbol de decisión.

Recordemos el flujo de trabajo de ML en Scikit-Learn: creamos el modelo / entrenamos / predecimos.

```python
from sklearn.tree import DecisionTreeClassifier

# Instanciamos un objeto de la clase DecisionTreeClassifier

clf = DecisionTreeClassifier(max_depth = 5, random_state = 42) 
```

```python
# Entrenamos el modelo

clf.fit(X.values,y.values)
```

```python
# Exploramos algunas de las características del modelo entrenado. Para pensar, 
# ¿qúe representa cada uno de estos valores?

print(clf.classes_)
print(clf.n_classes_)
print(clf.max_features_)
print(clf.feature_importances_)
```

```python
# Predecimos
oneh = OneHotEncoder(handle_unknown="ignore")
oneh.fit(X[X.columns])
X_test = oneh.transform(X_test[X_test.columns])
y_pred = clf.predict(X_test.values)
```
